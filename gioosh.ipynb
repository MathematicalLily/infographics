```python
plt.plot(performance_products_quality['Conversions'], label='conversions')
plt.plot(performance_products_quality['Clicks'], label='clicks')
plt.plot(performance_products_quality['Impressions'], label='impressions')
plt.plot(performance_products_quality['Total Cost'], label='total costs')
plt.plot(performance_products_quality['Total Revenue'], label='total revenue')

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Factors Influencing Revenue')
# Displaying the plot
plt.show()


# plt.plot(x2, y2, label='Line 2')
# plt.plot(x3, y3, label='Line 3')
```

```python
plt.plot(performance_products_quality['Total Cost'], label='total costs')
plt.plot(performance_products_quality['Total Revenue'], label='total revenue')
plt.plot(performance_products_quality['Total Revenue'] - performance_products_quality['Total Cost'], label='Profit')
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Profitability Measures')
```
